CREATING A FUNCTION WITH ROWS AND COLUMNS FOR CREATING A DATAFRAME

```python
def create_df(nRows,nColumns):
  arr=np.random.randint(0,10,(nRows,nColumns))
  df=pd.DataFrame(arr)
  df.index=['R'+ str(x) for x in np.arange(1,nRows+1)]
  df.columns=['C' +str(x) for x in np.arange(1,nColumns+1)]
  return df
create_df(5,3)
```

Go through each row of the dataframe and delete it(drop) if the coulmn is null

```python
for r in df.index:
  for c in df.columns:
    if pd.isnull(df.loc[r,c]):
      df.drop(r,inplace=True)
      break
      
```

now the empty values have been removed

```python
df.describe()
```

Another way of removing the rows which have empty data is using iterrows

```python
for i,c in df.iterrows():
  print(i)
  print(c)
  break
```

using iterrows we can get the index and each of the row and any funtion returns if anyone of the part is true

```python
for i,r in df.iterrows():
  if pd.isnull(r).any():
    df.drop(i,inplace=True)
    
```

```python
df.describe()
```

One more way for deleting the null values in the dataframe

```python
df.dropna(inplace=True)
```

```python
per_75=df.distance.quantile(0.75)
```

```python
df_=df.copy()
df_=df_[
   (df_['year']>=2010) &
   ((df_['method']=='Radial Velocity') | (df_['method']=='Transit')) &
   (df_['distance'] >per_75)
]
```
